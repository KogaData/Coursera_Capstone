- ### Clustering
Clustering is one of the most exploratory data analysis techniques used to get an understanding about the structure of the data. It can be defined as the task of identifying subgroups in the data such that data points in the same subgroup (cluster) are very similar while data points in different clusters are very dissimilar. Clustering is an unsupervised machine learning techniques, where there are no defined dependent and independent variables. The patterns in the data are used to identify or group similar observations. <br>
The objective of the clustering algorithm is to ensure that the distance between data points in a cluster is very low compared to the distance between the clusters. We used the K-Means clustering for creating neighborhood segmentation because the technique is simple to implement. K-Means is an iterative clustering algorithms with a predetermined number of cluster points. The algorithm assigns each data point to one of the $K$ clusters based on the feature similarity, measured on a metric that minimizes the sum of squared distances between the cluster centroids and its associated data points, that is:<br>     $$Minimize \sum^{K}_{j=1}{\sum^{N}_{i=1}{(x_{ij}-c_j)^2}}$$
where 
- $K=$ number of clusters,<br>
- $ N=$number of data points
- $c=$ centroind of cluster $j$
The main input for K-Means clustering is the number of clusters, which we derived using the concept of minimizing within cluster sum of squares (wcss), popularly known as the elbow method.
