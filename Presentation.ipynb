## Discussions

A SWOT (strength, weaknesses, opportunities,threats) analysis looks at a company's internal strengths and weaknesses and the external opportunities and threats within the market. With this analysis, we can easily highlight the direct areas of opportunity EMKO holdings can start building on or work to overcome. Using the cluster results, we are able to derive EMKO holdings' SWOT analysis chart.
- strengths: EMKO 's financial strength, skills set, reputation in Canada etc
-  Weaknesses: internal faults within the company's control
- Opportunities: requires looking at the banking industry, market and world as a whole to identify chances for a company to grow the business
- Threats: like opportunities, exist outside the company, but can be damaging to its success and profitability. We consider government regulation, influence of other companies, the economy, consumer trends, and more.
